```python
df=pd.read_csv("colleges.csv")
```

```python
df
```

# Removing NaN values

```python
df["rating"]=df["rating"].fillna(df["rating"].median())
```

```python

df
```

```python

df=df.replace("– / –",0)
```

```python

df["courses"] =  df[["courses"]].replace("1 Course",1)
```

```python

df
```

```python
df["courses"].info()
```

```python

df["courses"]=df["courses"].astype(int)
```

```python

df[['low_pkg', 'high_pkg']] = df['package'].str.split('-', expand=True)
df[["low_fee","high_fee"]]=df["fees_range"].str.split("-", expand=True)
```

```python

df
```

```python

df=df.fillna(0)
```

```python
df = df.replace('LPA', '', regex=True)
```

```python
df
```

```python

df[[ "low_pkg","high_pkg"]]=df[[ "low_pkg","high_pkg"]].astype(float)
```

```python

df[[ "low_fee","high_fee"]].info()
```

```python
df[[ "low_fee","high_fee"]] =  df[[ "low_fee","high_fee"]].replace(' Cr','0000000'  ,regex=True)
```

```python

df[[ "low_fee","high_fee"]]=df[[ "low_fee","high_fee"]].astype(float)
```

```python

df.head()
```

```python
df=df.drop(columns=["fees_range" ,"package"])
```

```python
df.head(1)
```

# Making new columns

```python
df["fees"]=( df["low_fee"] + df["high_fee"])/2
```

```python
df=df.drop(columns=["low_fee" ,"high_fee"])
```

```python

df
```

# Train Test splitting

```python
df.head()
```

```python
x =df[['fees','low_pkg','courses','high_pkg']]
```

```python
y= df['rating']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x, y , test_size =0.2)
```

```python
print(len(xtrain) ,len(ytrain), len(xtest) ,len(ytest))
 
```

# Applying Different models
- Linear Regression
- Decision Tree Regressor
- Random Forest Regressor
- XGBoost Regressor


<figure>
     <img src="https://static.javatpoint.com/tutorial/machine-learning/images/machine-learning-models.png"  style="width:8024px;height:510px;" >
</figure>

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(xtrain,ytrain)
model1.predict(xtest)
print("Traning = ",model1.score(xtrain,ytrain)*100)
print("Testing = ",model1.score(xtest,ytest)*100)
```

```python
from sklearn.tree import DecisionTreeRegressor
model2 = DecisionTreeRegressor()
model2.fit(xtrain,ytrain)
model2.predict(xtest)
print("Traning = ",model2.score(xtrain,ytrain)*100)
print("Testing = ",model2.score(xtest,ytest)*100)
```

```python
from sklearn.ensemble import RandomForestRegressor
model3 = RandomForestRegressor()
model3.fit(xtrain,ytrain)
model3.predict(xtest)
print("Traning = ",model3.score(xtrain,ytrain)*100)
print("Testing = ",model3.score(xtest,ytest)*100)
```

```python
from xgboost import XGBRegressor
model4 = XGBRegressor()
model4.fit(xtrain,ytrain)
model4.predict(xtest)
print("Traning = ",model4.score(xtrain,ytrain)*100)
print("Testing = ",model4.score(xtest,ytest)*100)
```
